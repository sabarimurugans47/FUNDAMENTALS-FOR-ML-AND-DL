```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt
```

#### Create 2D Dataset

```python
df=pd.DataFrame({
    'x1':[2,4,5,7],
    'x2':[3,5,8,10]
})
df
```

```python
df['x1'].mean(),df['x2'].mean()
```

#### Subtract samples with mean

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler(with_std=False)
```

```python
df_scaled=scaler.fit_transform(df)
df_scaled
```

```python
df_scaled[:,1:],df_scaled[:,:1]
```

```python
plt.scatter(df_scaled[:,1:],df_scaled[:,:1])
plt.show()
```

#### Implementing PCA technique

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(df_scaled)
```

#### Interpreting variance captured

```python
pca.explained_variance_ratio_ # First principle component captures most variance
```

```python
pca.components_ # Eigen vectors (Normalized)
```
