```python
import torch
import numpy as np

# Parameters
w = torch.tensor([1., 2.], requires_grad=True)
x = torch.tensor([[2., 3.]])
b = torch.tensor([2.], requires_grad=True)
y_true = torch.tensor([10.])
lr = 0.01  # Use Python float, not tensor

# Prediction
y_pred = x @ w + b

# Loss function (use PyTorch, not NumPy)
def loss(y, y_pred):
    return torch.mean((y - y_pred)**2)

# Gradient descent step
def gradient_opt(y, y_pred):
    global w, b
    l = loss(y, y_pred)         # compute loss
    l.backward()                # backpropagate

    with torch.no_grad():       # update weights
        w -= lr * w.grad
        b -= lr * b.grad

    w.grad.zero_()              # reset gradients
    b.grad.zero_()

    return l.item()
```

```python
gradient_opt(y_true,y_pred)
```
