# **HISTOGRAM**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

```python
heights=[150,150,151,150,152,150, 152, 155, 157, 160, 160, 162,
      165, 165, 167, 170, 172, 175, 177, 180]
sns.histplot(heights,bins=5,color='orange',stat='count',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Height')
plt.ylabel('count')
plt.show()
```

# **KERNEL DENSITY ESTIMATION**

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(heights, shade=True, color='coral')
plt.title('Height Distribution - KDE')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# **CUMMULATIVE DISTRIBUTION FUNCTION**

```python
heights_sorted = np.sort(heights)
n = len(heights_sorted)
cdf = np.arange(1, n+1) / n
plt.figure(figsize=(10, 5))
plt.plot(heights_sorted, cdf, marker='o', linestyle='-', color='red')
plt.title("Empirical CDF of Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

# **MEASURES OF CENTRAL TENDANCY (MEAN)**

mean.png

```python
mean=sum(heights)/len(heights)
print(mean)
```

# **MEASURES OF DISPERSION (VARIANCE)**

variance.png

```python
variance=sum((i-mean)**2 for i in heights)/len(heights)
print(variance)
```

# **STANDARD DEVIATION**

standard deviation.png

```python
std=np.sqrt(sum((i-mean)**2 for i in heights)/len(heights))
print(std)
```

# **NORMAL DISTRIBUTION (GAUSSIAN DISTRIBUTION)**

y=exp(-(x^2))
if x = 1 , y=exp(-(1)^2) y=0.3679 .
if x = 2 , y=exp(-(2)^2) y=0.0183 .
if x increases , y decreases

```python
from scipy.stats import norm
x = [-3, -2.5, -2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2, 2.5, 3]
mean = 0
std_dev = 1
y = [norm.pdf(i, mean, std_dev) for i in x]
plt.plot(x, y, marker='o', color='darkorange')
plt.title('Bell Curve: Normal Distribution (mean=0, std=1)')
plt.xlabel('x values')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
