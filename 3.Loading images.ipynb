```python
from matplotlib import image as jpg
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
img=jpg.imread('/content/image.jpg')
```

```python
plt.imshow(img)
type(img)#RGB
```

```python
import cv2
img2=cv2.imread('/content/image.jpg')
plt.imshow(img2)
type(img2)#BGR
```

```python
#To convert BGR to RGB
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
from PIL import Image
img3=Image.open('/content/image.jpg')
plt.imshow(img3)
type(img3)
```

```python
#to convert jpegimage to array type
img3=np.array(img3)
type(img3)
```

```python
#to save and load
np.save('img.npy',img3)
```

```python
#to load the image
img4=np.load('img.npy')
plt.imshow(img4)
```
