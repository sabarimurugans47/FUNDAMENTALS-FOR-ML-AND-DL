# **MEASURING SKEWNESS**

```python
import numpy as np

# Data
data = [10, 15, 20, 25, 30, 35, 40, 45]
x = np.array(data)
n = len(x)

# Mean
mean_x = np.mean(x)

# Variance part (denominator base)
var = (1/(n-1)) * np.sum((x - mean_x)**2)

if var == 0:
    print(" Skewness undefined (variance is zero)")
else:
    # Numerator: average of cubed deviations
    numerator = (1/n) * np.sum((x - mean_x)**3)

    # Denominator: variance^(3/2)
    denominator = var**1.5

    skewness = numerator / denominator

    print(f" Skewness: {skewness:.4f}")
    if skewness > 0:
        print(" Right skewed (tail on right)")
    elif skewness < 0:
        print(" Left skewed (tail on left)")
    else:
        print(" Symmetric distribution")
```

```python
import numpy as np

data = [55, 60, 61, 65, 66, 67, 68, 70]
x = np.array(data)
n = len(x)
mean_x = np.mean(x)

# Variance part
var = (1/(n-1)) * np.sum((x - mean_x)**2)

if var == 0:
    print("Kurtosis undefined (variance is zero)")
else:
    # Numerator: average of deviations^4
    numerator = (1/n) * np.sum((x - mean_x)**4)

    # Denominator: variance^2
    denominator = var**2

    kurtosis = numerator / denominator

    print(f" Kurtosis: {kurtosis:.4f}")
```
